```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = "darkgrid")

%matplotlib inline
```

```python
# dataset seaborn에서 가지고 있는 dummydata
tips = sns.load_dataset("tips")
tips
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
#hue 색상
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="size")
```

```python
#style & hue
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker" , style="size")
```

```python
#그래프 색상 변경
sns.relplot(x="total_bill",y="tip",data=tips,hue="size" , palette="Set2")
```

```python
#마커된 크기 변경
sns.relplot(x="total_bill",y="tip",data=tips,size="size")
```

```python
#임의의 데이터 생성 Linear Gradient Graphe
df = pd.DataFrame(dict(time=np.arange(500),value=np.random.randn(500).cumsum()))
df
```

```python
sns.relplot(x="time" , y ="value",data=df,kind="line")
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.relplot(x="timepoint",y="signal",data=fmri,kind="line")
```

```python
#테두리 없에고 싶을때 ci = None
sns.relplot(x="timepoint",y="signal",data=fmri,kind="line",ci=None)
```

```python
sns.relplot(x="timepoint",y="signal",data=fmri,kind="line",ci="sd")
```

```python
sns.relplot(x="timepoint",y="signal",data=fmri,kind="line",ci=None,hue="region",style="event",palette="Greens")
```

```python
#cat plot => 서로의 범주를 나눔 , 카테소리로 분류하여 , 막대 그래프
#rel => 관계형 데이터 사용 , 시간에 따른 변화
sns.catplot(x="day",y="total_bill",data=tips) # 요일별 요금 정산
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="box") # 요일별 요금 정산
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="violin",inner="stick",split=True,hue="sex") # 요일별 요금 정산
```
