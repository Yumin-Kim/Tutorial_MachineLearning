```python
m = folium.Map(location=[35.7743481,129.3099905],zoom_start=100)
m
```

```python
m = folium.Map(location=[35.7743481,129.3099905],zoom_start=50,tiles="OpenStreetMap")
m
```

```python
#마커 그리기
tooltip = "클릭해주세요"
folium.Marker([35.7743481,129.3099905],popup="<i>위</i>",tooltip=tooltip ).add_to(m)
folium.Marker([35.7243481,129.3099905],popup="<i>위</i>",tooltip=tooltip ).add_to(m)
folium.Marker([35.7343481,129.3099905],popup="<i>위</i>",tooltip=tooltip ).add_to(m)
m
```

```python
# icon && color
folium.Marker([35.7743481,129.3099905],icon=folium.Icon(icon="cloud")).add_to(m)
folium.Marker([35.7243481,129.3099905],icon=folium.Icon(icon="ad",color="green")).add_to(m)
m
```

```python
folium.Circle([35.7743481,129.3099905],radius=2000,color="red",fill=True,fill_color="red").add_to(m)
m
```

```python
m.add_child(folium.ClickForMarker(popup="Click"))
m
```

```python
m.add_child(folium.LatLngPopup())
m
```
