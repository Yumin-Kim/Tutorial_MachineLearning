```python
import tensorflow.compat.v1 as tf
import numpy as np
tf.disable_v2_behavior()
node1 = tf.constant("Hello")
print("node1",node1)
sess = tf.Session()
out = sess.run(node1)
print(out)
```

```python
node1 = tf.placeholder(tf.float32)
node2 = tf.placeholder(tf.float32)
adder_node = node1 + node2
print(sess.run(adder_node , feed_dict={node1:10 ,node2 : 20}))
print(sess.run(adder_node , feed_dict={node1:[1,3] , node2 : [2,6]}))
```

```python
x_train = [1,2,3]
y_train = [1,2,3]

W = tf.Variable(tf.random_normal([1]) , name = 'weight')
b = tf.Variable(tf.random_normal([1]) , name = 'bias')
hypothesis = x_train * W + b
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
'''
train = optimizer.minimize(cost) 그래프에서의 루트 노드라고 비유!!
'''
```

```python
sess.run(tf.global_variables_initializer())
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step,sess.run(cost),sess.run(W),sess.run(b))
```
